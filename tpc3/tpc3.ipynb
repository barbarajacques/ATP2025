```python
import random


def criaLista(N):
    #inicializar uma lista
    res=[]
    i=0
    for i in range(1,N+1,1):
        res.append(random.randint(1,100))
    return res
def leLista(N):
    res=[]
    for i in range (N):
        num=int(input(f"introduza o número {i+1}/{N})"))
        res.append(num)
    return res
def somaLista(lista):
    soma=0
    for i in lista:
        soma+=i
    return soma
def mediaLista(lista):
    if len(lista)==0:
        return 0
    soma=somaLista(lista)
    return soma/len(lista)
def maiorLista(lista):
    maior=lista[0]
    for elem in lista[1:]:
        if elem>maior:
            maior=elem
    return maior
def menorLista(lista):
    menor=lista[0]
    for elem in lista[1:]:
        if elem<menor:
            menor=elem
    return menor
def estaOrdenadaCrescente(lista):
    #Verifica se a lista está ordenada por ordem crescente
    return lista==sorted(lista)

def estaOrdenadaDecrescente(lista):
    #Verifica se a lista está ordenada por ordem decrescente
    return lista==sorted(lista, reverse=True)

def procuraElemento(lista, valor):
    #Procura um elemento na lista e devolve a posição (ou -1 se não existir)
    if valor in lista:
        return lista.index(valor)
    else:
        return -1
lista=[]

while True:
    print("\n--- MENU ---")
    print("(1) Criar Lista (aleatória)")
    print("(2) Ler Lista (introduzida pelo utilizador)")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Está ordenada por ordem crescente?")
    print("(8) Está ordenada por ordem decrescente?")
    print("(9) Procurar um elemento")
    print("(0) Sair")

    opcao=input("Escolha uma opção: ")

    if opcao=='0':
        print("A lista atual é:", lista)
        print("Programa terminado.")
        break

    elif opcao=='1':
        N = int(input("Quantos elementos pretende criar? "))
        lista = criaLista(N)
        print("Lista criada:", lista)

    elif opcao=='2':
        N=int(input("Quantos elementos pretende introduzir? "))
        lista=leLista(N)
        print("Lista introduzida:", lista)

    elif opcao=='3':
        print("Soma dos elementos:", somaLista(lista))

    elif opcao=='4':
        print("Média dos elementos:", mediaLista(lista))

    elif opcao=='5':
        print("Maior elemento:", maiorLista(lista))

    elif opcao=='6':
        print("Menor elemento:", menorLista(lista))

    elif opcao=='7':
        print("Ordenada por ordem crescente?", "Sim" if estaOrdenadaCrescente(lista) else "Não")

    elif opcao=='8':
        print("Ordenada por ordem decrescente?", "Sim" if estaOrdenadaDecrescente(lista) else "Não")

    elif opcao=='9':
        valor=int(input("Introduza o valor a procurar: "))
        pos=procuraElemento(lista, valor)
        if pos!=-1:
            print(f"O valor {valor} encontra-se na posição {pos}.")
        else:
            print("O valor não foi encontrado na lista.")

    else:
        print("Opção inválida. Tente novamente.")
```
